```python
#export dataframe to json file

import json

# group questions and answers by label
tmp = (df.groupby("label").apply(
    
    lambda x: dict(patterns = x["Pytanie"].to_list(), 
                   responses = x["Odpowiedź"].to_list())).to_dict())

out = {"intents": []}

for k, v in tmp.items():
    out["intents"].append({})
    out["intents"][-1]["tag"] = k
    out["intents"][-1]["patterns"] = v["patterns"]
    out["intents"][-1]["responses"] = v["responses"]

# print(json.dumps(out, indent=4))
```

```python
# save to json file
with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(out, f, indent=4, ensure_ascii=False)
```
